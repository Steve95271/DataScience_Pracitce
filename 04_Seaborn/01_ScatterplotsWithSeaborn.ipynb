```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
df.head(10)
```

```python
plt.figure(dpi=127)
sns.scatterplot(
    data=df,
    x="sales",
    y="salary",
    hue="salary"
)
plt.savefig("sales_salary_plot.jpg")
```

**Choosing palette that change the color of the data**

https://matplotlib.org/tutorials/colors/colormaps.html

```python
plt.figure(dpi=127)
sns.scatterplot(
    data=df,
    x="sales",
    y="salary",
    hue="salary",
    palette="viridis"
)
```
