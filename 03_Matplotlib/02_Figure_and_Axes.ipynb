```python
import matplotlib.pyplot as plt;
```

```python
import numpy as np
```

```python
a = np.linspace(0, 10, 11)
a
```

```python
b = pow(a, 4)
```

```python
b
```

```python
x = np.arange(0, 10)
x
```

```python
y = x * 2
y
```

```python
fig = plt.figure()
# 0,0 means bottom left value
# 1,1 means 100% height and 100% width
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_title("Power of 4")
axes1.set_xlabel("A")
axes1.set_ylabel("B")


axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.set_xlim(4,6)
axes2.set_ylim(0,2000)
axes2.set_title("Zoom in")
axes2.plot(a,b)
```

```python
# figsize 调整比例和大小，dpi 调整图片的大小，与清晰度有关
fig = plt.figure(figsize=(4,3), dpi=200)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
```
