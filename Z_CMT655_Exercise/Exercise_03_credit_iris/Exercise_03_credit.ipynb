In this exercise, you will explore the [German Credit Dataset](https://archive.ics.uci.edu/dataset/144/statlog+german+credit+data).

The dataset contains data of 1000 German customer profiles described by 20 attributes in the 1990s. Each instance has a rating that can be good (1) or bad (2).
The features are

- Variable 1: Status of existing checking account
  - A11 :      ... <    0 DM
  - A12 : 0 <= ... <  200 DM
  - A13 :      ... >= 200 DM /salary assignments for at least 1 year
  - A14 : no checking account

- Variable 2: Duration in month

- Variable 3: Credit history
  - A30 : no credits taken/
		    all credits paid back duly
  - A31 : all credits at this bank paid back duly
  - A32 : existing credits paid back duly till now
  - A33 : delay in paying off in the past
  - A34 : critical account/
		    other credits existing (not at this bank)

- Variable 4: Purpose
  - A40 : car (new)
  - A41 : car (used)
  - A42 : furniture/equipment
  - A43 : radio/television
  - A44 : domestic appliances
  - A45 : repairs
  - A46 : education
  - A47 : (vacation - does not exist?)
  - A48 : retraining
  - A49 : business
  - A410 : others

Variable 5: Credit amount

Variable 6: Savings account/bonds
  - A61 :          ... <  100 DM
  - A62 :   100 <= ... <  500 DM
  - A63 :   500 <= ... < 1000 DM
  - A64 :          .. >= 1000 DM
  - A65 :   unknown/ no savings account

Variable 7: Present employment since
  - A71 : unemployed
  - A72 :       ... < 1 year
  - A73 : 1  <= ... < 4 years  
  - A74 : 4  <= ... < 7 years
  - A75 :       .. >= 7 years

Variable 8: Installment rate in percentage of disposable income

Variable 9: Personal status and sex
  - A91 : male   : divorced/separated
  - A92 : female : divorced/separated/married
  - A93 : male   : single
  - A94 : male   : married/widowed
  - A95 : female : single

Variable 10: Other debtors / guarantors
  - A101 : none
  - A102 : co-applicant
  - A103 : guarantor

Variable 11: Present residence since

Variable 12: Property
  - A121 : real estate
  - A122 : if not A121 : building society savings agreement/
				   life insurance
  - A123 : if not A121/A122 : car or other, not in attribute 6
  - A124 : unknown / no property

Variable 13: Age in years

Variable 14: Other installment plans 
  - A141 : bank
  - A142 : stores
  - A143 : none

Variable 15: Housing
  - A151 : rent
  - A152 : own
  - A153 : for free

Variable 16: Number of existing credits at this bank

Variable 17: Job
  - A171 : unemployed/ unskilled  - non-resident
  - A172 : unskilled - resident
  - A173 : skilled employee / official
  - A174 : management/ self-employed/
		     highly qualified employee/ officer

Variable 18: Number of people being liable to provide maintenance for

Variable 19: Telephone
  - A191 : none
  - A192 : yes, registered under the customers name

Variable 20: foreign worker
  - A201 : yes
  - A202 : no
