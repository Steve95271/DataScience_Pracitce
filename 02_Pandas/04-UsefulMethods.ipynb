```python
import pandas as pd
import numpy as np
```

```python
tips_df = pd.read_csv("tips.csv")
```

```python
def yelp(price):
    if price < 10:
        return '$'
    elif price >= 10 and price < 30:
        return '$$'
    else:
        return '$$$'
```

```python
tips_df['Expensive'] = tips_df['total_bill'].apply(yelp)
```

```python
tips_df['total_bill'].apply(lambda bill : bill * 0.25)
```

```python
def quality(total_bill, tip):
    if tip / total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
```

```python
tips_df.head()
```

```python
tips_df['tip_quality'] = (tips_df[['total_bill', 'tip']]
    .apply(
        lambda tips_df : quality(tips_df['total_bill'], tips_df['tip']),
        axis=1
    )
)
```

```python
tips_df
```

Using NumPy vectorize() function will run faster

```python
tips_df['tip_quality'] = np.vectorize(quality)(tips_df['total_bill'], tips_df['tip'])
```

```python
import timeit

# code snippet to be executed only once
setup = '''
import numpy as np
import 02_Pandas as pd
df = pd.read_csv('tips.csv')
def quality(total_bill,tip):
    if tip/total_bill  > 0.25:
        return "Generous"
    else:
        return "Other"
'''

# code snippet whose execution time is to be measured
stmt_one = '''
df['Tip Quality'] = df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1)
'''

stmt_two = '''
df['Tip Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
'''
```

```python
timeit.timeit(
    setup = setup,
    stmt = stmt_one,
    number = 5000
)
```

```python
timeit.timeit(
    setup = setup,
    stmt = stmt_two,
    number = 5000
)
```
